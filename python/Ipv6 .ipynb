```python
# I scrape data using this 
import requests
import urllib.request
import time
from bs4 import BeautifulSoup

# Set the URL you want to webscrape from
url = 'http://bgp.potaroo.net/stats/nro/archive'

# Connect to the URL
response = requests.get(url)
# Parse HTML and save to BeautifulSoup objectÂ¶
soup = BeautifulSoup(response.text, "html.parser")
for i in range(1500,1999): #'a' tags are for links
    one_a_tag = soup.findAll('a')[i]
    link = one_a_tag['href']
    download_url = 'http://bgp.potaroo.net/stats/nro/archive/'+link
    urllib.request.urlretrieve(download_url,'./'+link[link.find('/turnstile_')+1:]+'.txt') 
    time.sleep(1)
    
```
